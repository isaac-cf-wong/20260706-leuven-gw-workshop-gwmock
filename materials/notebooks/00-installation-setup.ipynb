## 1. The ecosystem — four packages

| Package | Role | You use it for |
|---|---|---|
| **`gwmock`** | Umbrella / orchestrator | YAML config → full MDC (CLI) |
| **`gwmock-pop`** | Population sampling (JAX) | source catalogues: BBH/BNS/NSBH parameters |
| **`gwmock-signal`** | Signal simulation | waveforms → detector projection → injection, SGWB |
| **`gwmock-noise`** | Noise simulation | coloured/correlated noise, lines, glitches |

`pip install gwmock` pulls in all four. Each of the three engines is an
independent library you can also use directly in Python — that is exactly
what notebooks 02–04 do.

```python
import gwmock, gwmock_pop, gwmock_signal, gwmock_noise

for pkg in (gwmock, gwmock_pop, gwmock_signal, gwmock_noise):
    print(f"{pkg.__name__:15s} {pkg.__version__}")
```

## 2. The CLI in one table

| Command | What it does |
|---|---|
| `gwmock simulate config.yaml` | run a simulation (re-run resumes from checkpoint) |
| `gwmock config --list` / `--get <label>` | browse / copy bundled example configs |
| `gwmock config --interactive` | TUI config editor: live preview, `/`-commands, script generation |
| `gwmock validate` | check outputs against metadata hashes |
| `gwmock merge` | combine frame files (e.g. signal + noise) |
| `gwmock batch` | prepare/submit Slurm batch jobs |
| `gwmock repository` | publish/verify datasets on Zenodo |

There are also `gwmock-pop` and `gwmock-noise` CLIs for the engines.
